### Filters

```python
"""
CLIENTS DATA FRAME
Filters
1. Contracts from 2015 onwards
"""

clients_df_filtered = clients_df.copy()
date_threshold_2015 = datetime(2014,12,31)
clients_df_filtered.drop(clients_df[clients_df["application_date"] <= date_threshold_2015].index, inplace = True)
clients_df_filtered
```

```python
"""
2. For a client to be eligible, it must have at least two years of information within the company 
(even if it canceled the product).
Data delivered on 30/11/2019 (day/month/year).
"""
date_threshold_2 = datetime(2017,11,30)
clients_df_filtered.drop(clients_df_filtered[clients_df_filtered["application_date"] > date_threshold_2].index, inplace=True)
clients_df_filtered
```

```python
"""
3. Take out clients with more than 75% of their info missing.
I didn't consider customerId as a relevant info, so there are 9 columns with relevant info
Rows with null values > 7 will be droped
"""
index_75 = clients_df_filtered.isnull().sum(axis=1).index[clients_df_filtered.isnull().sum(axis=1) >= 7]
clients_df_filtered.drop(index_75, inplace=True)
clients_df_filtered
```

```python
"""Duplicates"""
clients_df_filtered.drop_duplicates(inplace=True)
clients_df_filtered
```

```python
"""
PRODUCTS DATA FRAME
"""

products_df_filtered = products_df.copy()
index_filtered = products_df[products_df.CustomerId.isin(clients_df_filtered.CustomerId) == False].index
products_df_filtered.drop(index_filtered, inplace=True)
products_df_filtered.drop(["ContractId"], axis = 1, inplace=True)
products_df_filtered
```

```python
products_df_filtered = products_df_filtered.groupby(["CustomerId"]).count()
products_df_filtered
```

```python
"""
TRANSACTIONS DATA FRAME
"""
transactions_df_filtered = transactions_df.copy()
index_filtered_1 = transactions_df[transactions_df.CustomerId.isin(clients_df_filtered.CustomerId) == False].index
transactions_df_filtered.drop(index_filtered_1, inplace=True)
transactions_df_filtered.drop(["Transaction"], axis=1, inplace=True)
transactions_df_filtered

```

```python
transactions_df_filtered = transactions_df_filtered.groupby(["CustomerId"]).sum()
transactions_df_filtered
```

```python
"""
SCORE DATA FRAME
"""
score_df_filtered = score_df.copy()
index_filtered_2 = score_df[score_df.CustomerId.isin(clients_df_filtered.CustomerId) == False].index
score_df_filtered.drop(index_filtered_2, inplace=True)
score_df_filtered
```

```python
clients_app_date = clients_df_filtered[["CustomerId", "application_date"]].copy()
score_app_date = pd.merge(score_df_filtered, clients_app_date, how="inner", on="CustomerId")
index_filtered_3 = score_app_date[
    (score_app_date["Date"].dt.month == score_app_date["application_date"].dt.month) == False].index
score_app_date.drop(index_filtered_3, inplace=True)
score_app_date
```

```python
score_app_date[["CustomerId", "Score"]]
```

### Final Dataset

```python
data = pd.merge(clients_df_filtered, products_df_filtered, how = "inner", on = "CustomerId")
data = pd.merge(data, transactions_df_filtered, how="inner", on = "CustomerId")
data = pd.merge(data, score_app_date[["CustomerId", "Score"]], how="inner", on="CustomerId")
data
```

```python
"""
Add age, target
Drop columns not considered for the model
"""
data["age"] = (data.application_date - data.birth_date).dt.days // 365
mask = (data["exit_date"] - data["application_date"]).dt.days > 730
data['target'] = np.where(mask, 1, 0)
data.loc[data.exit_date.isnull(), "target"] = 1
data.rename(columns={"Value":"balance"}, inplace=True)

data.drop(["CustomerId", "Surname", "exit_date", "birth_date"], axis=1, inplace=True)
data
```

#### Label and One-hot Encoding

```python
data_normalized["application_date"] = data_normalized.application_date.dt.month
data_normalized
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data_label_encoded = data_normalized.copy()
keys_for_label = ['Geography', 'Gender']

for key in keys_for_label:
    data_label_encoded [key] = labelencoder.fit_transform(data_label_encoded[key])
```

```python
data_oneH_encoded = data_normalized.copy()
for key in keys_for_label:
    data_oneH_encoded = pd.get_dummies(data_oneH_encoded, prefix=[key], columns=[key])
```

```python
data_0 = data_label_encoded[data_label_encoded.target == 0]
data_1 = data_label_encoded[data_label_encoded.target == 1].sample(frac=0.5)
frames = [data_0, data_1]
data_balanced_label_encoded = pd.concat(frames)

data_oh_0 = data_oneH_encoded[data_oneH_encoded.target == 0]
data_oh_1 = data_oneH_encoded[data_oneH_encoded.target == 1].sample(frac=0.5)
frames = [data_oh_0,data_oh_1]
data_balanced_one_hot = pd.concat(frames)
```

```python
data_balanced_label_encoded.groupby("target").size().plot.pie(ylabel = "Target", label="asdsd", figsize=(5,5))
```

### Models

```python
"""LABEL ENCODED"""
#X_data = data_label_encoded.drop(["target"], axis=1)
#y_data = data_label_encoded.target

"""LABEL BALANCE ENCODED"""
#X_data = data_balanced_label_encoded.drop(["target"], axis=1)
#y_data = data_balanced_label_encoded.target

"""LABEL ONE-HOT-ENCODED"""
#X_data = data_oneH_encoded.drop(["target"], axis=1)
#y_data = data_oneH_encoded.target

"""LABEL BALANCE ONE-HOT-ENCODED"""
X_data = data_balanced_one_hot.drop(["target"], axis=1)
y_data = data_balanced_one_hot.target

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.35, random_state=0)
```

```python
"""DecisionTreeClassifier"""
from sklearn.tree import DecisionTreeClassifier
cl_tree = DecisionTreeClassifier(criterion="entropy", max_depth=15)
cl_tree.fit(X_train,y_train)
cl_tree.score(X_test, y_test)
```

```python
"""RandomForestClassifier"""
from sklearn.ensemble import RandomForestClassifier
clf_RFC = RandomForestClassifier(max_depth=2, random_state=0)
clf_RFC.fit(X_train, y_train)
clf_RFC.score(X_test, y_test)
```

```python
"""XGBOOST CLASSIFIER"""
import xgboost as xgb
clf_XGB_0 = xgb.XGBClassifier()
clf_XGB_0.fit(X_train,y_train)
clf_XGB_0.score(X_test,y_test)
```

#### clf_XGB_0

```python

c_parameter_name = 'learning_rate'
c_parameter_values = [0.01, 0.1, 0.2, 0.3, 0.5]
c_best_parameter = 0
c_best_accuracy = 0
c_worst_parameter = 0
c_worst_accuracy = 100
df_1 = pd.DataFrame(columns=[c_parameter_name, 'accuracy'])
for input_parameter in c_parameter_values:
    clf_XGB_0 = xgb.XGBClassifier(learning_rate = input_parameter)
    
    
    clf_XGB_0.fit(X_train, y_train)
    y_pred = clf_XGB_0.predict(X_test)
    
    acc_score = accuracy_score(y_test,y_pred)*100
    
    df_1 = df_1.append({c_parameter_name : input_parameter , 'accuracy' : acc_score}, ignore_index=True)
    
    if acc_score > c_best_accuracy:
        c_best_accuracy = acc_score
        c_best_parameter = input_parameter
        c_best_model_x0 = clf_XGB_0
        
    if acc_score < c_worst_accuracy:
        c_worst_accuracy = acc_score
        c_worst_parameter = input_parameter
        c_worst_model_x0 = clf_XGB_0
    
print(df_1)
print("")
plt.figure(figsize=(12,6))
sbn.pointplot(x=c_parameter_name, y="accuracy", data=df_1)
title = 'clf_XGB_0 Accuracy(%) vs ' + c_parameter_name + ' parameter'
plt.title(title)
plt.xticks(rotation= 90)
plt.grid()
```

```python
clf_XGB_1 = xgb.XGBClassifier(learning_rate=0.1, max_depth = 10)
clf_XGB_1.fit(X_train,y_train)
clf_XGB_1.score(X_train,y_train)
```

```python
c_parameter_name = 'max_depth'
c_parameter_values = [2,4,6,8]
c_best_parameter = 0
c_best_accuracy = 0
c_worst_parameter = 0
c_worst_accuracy = 100
df_2 = pd.DataFrame(columns=[c_parameter_name, 'accuracy'])
for input_parameter in c_parameter_values:
    clf_XGB_1 = xgb.XGBClassifier(max_depth = input_parameter)
    
    
    clf_XGB_1.fit(X_train, y_train)
    y_pred = clf_XGB_1.predict(X_test)
    
    acc_score = accuracy_score(y_test,y_pred)*100
    
    df_2 = df_2.append({c_parameter_name : input_parameter , 'accuracy' : acc_score}, ignore_index=True)
    
    if acc_score > c_best_accuracy:
        c_best_accuracy = acc_score
        c_best_parameter = input_parameter
        c_best_model_x1 = clf_XGB_1
        
    if acc_score < c_worst_accuracy:
        c_worst_accuracy = acc_score
        c_worst_parameter = input_parameter
        c_worst_model_x1 = clf_XGB_1
    
print(df_2)
print("")
plt.figure(figsize=(12,6))
sbn.pointplot(x=c_parameter_name, y="accuracy", data=df_2)
title = 'clf_XGB_1 Accuracy(%) vs ' + c_parameter_name + ' parameter'
plt.title(title)
plt.xticks(rotation= 90)
plt.grid()
```
